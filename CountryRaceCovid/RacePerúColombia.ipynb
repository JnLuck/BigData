```python
# Importar pandas
import pandas as pd
```

```python
# Leer el archivo CSV
datos = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv")
```

```python
datos
```

```python
import bar_chart_race as bcr
```

```python
# Seleccionar solo las columnas necesarias
datos = datos[["location", "date", "new_cases_smoothed"]]

# Reemplazar NaN en 'new_cases_smoothed' con 0
datos["new_cases_smoothed"].fillna(0, inplace=True)

# Filtrar los datos para obtener solo los registros de Perú y Guatemala
paises = ["Peru", "Colombia"]
datos_filtrados = datos[datos["location"].isin(paises)]

# Convertir la columna 'date' a formato datetime
datos_filtrados["date"] = pd.to_datetime(datos_filtrados["date"])

# Crear una tabla pivote para el gráfico
pivot_data = datos_filtrados.pivot(index='date', columns='location', values='new_cases_smoothed')

# Rellenar los valores faltantes con 0
pivot_data = pivot_data.fillna(0)

# Crear el gráfico de carrera
bcr.bar_chart_race(
    df=pivot_data,
    filename='covid_cases_race3.mp4',
    title='New COVID-19 Cases Smoothed: Peru vs Colombia',
    period_length=500
)
```
