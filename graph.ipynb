```python
#Importamos la libreria matplotlib
import matplotlib.pyplot as plt
```

```python
#Importamos la libreria numpy
import numpy as np
```

# **GRÁFICO 1**

```python
# Configuramos el gráfico:
fig, graficos = plt.subplots(figsize=(6,6))

graficos.scatter(
    x=[1,2,3],
    y=[3,2,1]
)

plt.show() # Visualizamos el gráfico
```

# **GRÁFICO 2**

```python
# Configuramos el gráfico:
# Indicamos que la distribución de esos 2 gráficos será de 1 fila y 2 columnas
# Tamaño del gráfico será de 6X6 pulgadas
fig, graficos = plt.subplots(1,2,figsize=(8,6))

# Gráficamos "puntos" en las coordenadas X e Y en el primer gráfico [0]
graficos[0].scatter(
    x=[1,2,3],
    y=[3,2,1]
)

# Gráficamos "puntos" en las coordenadas X e Y en el primer gráfico [1]
graficos[1].scatter(
    x=[2,4,6,8,10],
    y=[4,16,35,64,100]
)

plt.show() # Visualizamos ambos gráficos
```

# **GRÁFICO 3**

```python
# Configuramos el gráfico:
# Indicamos que la distribución de esos 2 gráficos será de 1 fila y 2 columnas
# Tamaño del gráfico será de 6X6 pulgadas
fig, graficos = plt.subplots(1,2,figsize=(8,6))

# Gráficamos "puntos" en las coordenadas X e Y en el primer gráfico [0]
graficos[0].scatter(
    x=[1,2,3],
    y=[3,2,1]
)

eje_x = np.arange(1,10,1);
eje_y = []
for numero in eje_x:
  cuadrado = numero*numero
  eje_y.append(cuadrado)

# Gráficamos "puntos" en las coordenadas X e Y en el primer gráfico [1]
graficos[1].scatter(
    x = eje_x,
    y = eje_y
)

plt.show() # Visualizamos ambos gráficos
```
