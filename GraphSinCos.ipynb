```python
# Importar matplotlib para graficar y numpy
import matplotlib.pyplot as plt
import numpy as np
```

# **Gr√°fico Seno y Coseno (Lineas)**

```python
# Generar valores para el eje x de -10 a 10 con paso de 0.25
x = np.arange(20, 50, 0.1)

# Calcular los valores de seno y coseno correspondientes a cada valor de x
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura de tama√±o 10x6 pulgadas para el gr√°fico
plt.figure(figsize=(10, 6))

# Graficar la funci√≥n seno como puntos azules ('b') con etiqueta 'Seno'
plt.plot(x, y1, color='b', label='Seno')

# Graficar la funci√≥n coseno como puntos rojos ('r') con etiqueta 'Coseno'
plt.plot(x, y2, color='r', label='Coseno')

# Etiquetar el eje x como 'x'
plt.xlabel('x')

# Etiquetar el eje y como 'y'
plt.ylabel('y')

# Establecer el t√≠tulo del gr√°fico como 'Funciones Seno y Coseno'
plt.title('Gr√°fico de las Funciones Seno y Coseno')

# Mostrar la leyenda del gr√°fico
plt.legend()

# Establecer los l√≠mites del eje y de -2 a 2
plt.ylim(-2, 2)

# Mostrar una cuadr√≠cula en el gr√°fico
plt.grid(True)

# Mostrar el gr√°fico en pantalla
plt.show()
```

# **Gr√°fico Seno y Coseno (Puntos)**

```python
# Generar valores para el eje x de -10 a 10 con paso de 0.25
x = np.arange(-10, 10, 0.25)

# Calcular los valores de seno y coseno correspondientes a cada valor de x
y1 = np.sin(x)
y2 = np.cos(x)

# Crear una figura de tama√±o 10x6 pulgadas para el gr√°fico
plt.figure(figsize=(10, 6))

# Graficar la funci√≥n seno como puntos azules ('b') con etiqueta 'Seno'
plt.scatter(x, y1, color='b', label='Seno')

# Graficar la funci√≥n coseno como puntos rojos ('r') con etiqueta 'Coseno'
plt.scatter(x, y2, color='r', label='Coseno')

# Etiquetar el eje x como 'x'
plt.xlabel('x')

# Etiquetar el eje y como 'y'
plt.ylabel('y')

# Establecer el t√≠tulo del gr√°fico como 'Funciones Seno y Coseno'
plt.title('Funciones Seno y Coseno')

# Mostrar la leyenda del gr√°fico
plt.legend()

# Establecer los l√≠mites del eje y de -2 a 2
plt.ylim(-2, 2)

# Mostrar una cuadr√≠cula en el gr√°fico
plt.grid(True)

# Mostrar el gr√°fico en pantalla
plt.show()
```

# **Gr√°ficos: Seno, Coseno, Tangente y Cuadr√°tica**

```python
# Generar valores para el eje x de -10 a 10 con paso de 0.25
x = np.arange(-10, 10, 0.25)

# Calcular los valores de las funciones
y_seno = np.sin(x)
y_coseno = np.cos(x)
y_tangente = np.tan(x)
y_cuadratica = x**2

# Configurar el gr√°fico con un solo plano cartesiano
plt.figure(figsize=(12, 8))

# Graficar el seno
plt.subplot(2, 2, 1)
plt.plot(x, y_seno, label='Seno', color='b')
plt.legend()
plt.grid(True)
plt.title('Seno')

# Graficar el coseno
plt.subplot(2, 2, 2)
plt.plot(x, y_coseno, label='Coseno', color='r')
plt.legend()
plt.grid(True)
plt.title('Coseno')

# Graficar la tangente
plt.subplot(2, 2, 3)
plt.plot(x, y_tangente, label='Tangente', color='g')
plt.legend()
plt.grid(True)
plt.title('Tangente')

# Graficar la funci√≥n cuadr√°tica
plt.subplot(2, 2, 4)
plt.plot(x, y_cuadratica, label='Cuadr√°tica', color='purple')
plt.legend()
plt.grid(True)
plt.title('Cuadr√°tica')

# Ajustar espacio entre subgr√°ficos
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```

# **Gr√°ficos: Seno, Coseno, Tangente, Cuadr√°tica, Linea1 y Linea2**

```python
# Generar valores para el eje x de -10 a 10 con paso de 0.25
x = np.arange(-10, 10, 0.25)

# Calcular los valores de las funciones
y_seno = np.sin(x)
y_coseno = np.cos(x)
y_tangente = np.tan(x)
y_cuadratica = x**2

# Configurar el gr√°fico con un solo plano cartesiano
plt.figure(figsize=(12, 8))

# Graficar el seno
plt.subplot(3, 2, 1)
plt.plot(x, y_seno, label='Seno', color='b')
plt.legend()
plt.grid(True)
plt.title('Seno')

# Graficar el coseno
plt.subplot(3, 2, 2)
plt.plot(x, y_coseno, label='Coseno', color='r')
plt.legend()
plt.grid(True)
plt.title('Coseno')

# Graficar la tangente
plt.subplot(3, 2, 3)
plt.plot(x, y_tangente, label='Tangente', color='g')
plt.legend()
plt.grid(True)
plt.title('Tangente')

# Graficar la funci√≥n cuadr√°tica
plt.subplot(3, 2, 4)
plt.plot(x, y_cuadratica, label='Cuadr√°tica', color='purple')
plt.legend()
plt.grid(True)
plt.title('Cuadr√°tica')

# Graficar la primera recta en el quinto subgr√°fico
plt.subplot(3, 2, 5)
plt.plot(x, 2*x + 1, label='y = 2x + 1', color='pink')
plt.legend()
plt.grid(True)
plt.title('Recta 1')

# Graficar la segunda recta en el sexto subgr√°fico
plt.subplot(3, 2, 6)
plt.plot(x, -3*x + 5, label='y = -3x + 5', color='c')
plt.legend()
plt.grid(True)
plt.title('Recta 2')

# Ajustar espacio entre subgr√°ficos
plt.tight_layout()

# Mostrar el gr√°fico
plt.show()
```
