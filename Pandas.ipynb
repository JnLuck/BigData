```python
# Importar la librería pandas
import pandas as pd
```

```python
# Importar la librería matplotlib
import matplotlib.pyplot as plt
```

```python
# Leer un archivo CSV desde una URL
df = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vTKv5ve5cbcTUOygS_FEd-swTOTVxC4DldFq-T2W4nPW-XqMYK8rVrMn_RDwf_ujF8o_1YyHAL98OiO/pub?gid=1414694331&single=true&output=csv")
```

```python
# Ajustar la opción de ancho de visualización en pandas a 200
pd.set_option("display.width", 200)

# Imprimir el DataFrame 'df'
print(df)
```

```python
# Lista datos
df
```

```python
# Contar la cantidad de registros donde la columna UBIGEO
cantidad_21 = df[df['UBIGEO'] == 211101].shape[0]

# Mostrar la cantidad
print("Cantidad de registros en Juliaca:", cantidad_21)
```

```python
# Contar la cantidad de registros donde la columna UBIGEO
cantidad_20 = df[df['UBIGEO'] == 210101].shape[0]

# Mostrar la cantidad
print("Cantidad de registros en Puno:", cantidad_20)
```

```python
df_puno = df[df["UBIGEO"].astype(str).str.startswith('21')]

# Contar la cantidad de registros para Puno
cantidad_registros_puno = df_puno.shape[0]

print("La cantidad de registros para Puno es:", cantidad_registros_puno)
```

```python
# Filtrar el DataFrame para obtener solo los registros de Juliaca y Puno
df_juliaca_puno = df[df['UBIGEO'].isin([211101, 210101])]

# Contar la cantidad de registros para cada valor de UBIGEO
conteo_ubigeo = df_juliaca_puno['UBIGEO'].value_counts()

# Crear las etiquetas personalizadas
etiquetas = ['Juliaca (211101)', 'Puno (210101)']

# Crear el gráfico de pastel con las etiquetas personalizadas
plt.figure(figsize=(8, 6))
conteo_ubigeo.plot(kind='pie', autopct=lambda p: '{:.0f}'.format(p * sum(conteo_ubigeo) / 100), labels=etiquetas, colors=plt.cm.tab20.colors)
plt.title('Distribución de Registros por UBIGEO (Juliaca y Puno)')
plt.ylabel('')

# Agregar el total como texto
total_registros = sum(conteo_ubigeo)
plt.text(-1.5, -1.5, f'Total: {total_registros}', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```

```python
# Filtrar filas donde UBIGEO sea igual a 210101 (Puno) o 211101 (Juliaca)
df_puno_juliaca = df[(df["UBIGEO"] == 210101) | (df["UBIGEO"] == 211101)]

# Imprimir el DataFrame
print(df_juliaca_puno)
```

```python
# Filtrar filas para hablantes de quechua (idioma_ninez == 1)
i_quechua = df_puno_juliaca[df_puno_juliaca['idioma_ninez'] == 1]
print("Hablantes quechua: ", len(i_quechua))

# Filtrar filas para hablantes de aymara (idioma_ninez == 2)
i_aymara = df_puno_juliaca[df_puno_juliaca['idioma_ninez'] == 2]
print("Hablantes de aymara: ", len(i_aymara))

# Filtrar filas para hablantes de castellano (idioma_ninez == 4)
i_castellano = df_puno_juliaca[df_puno_juliaca['idioma_ninez'] == 4]
print("Hablantes de castellano: ", len(i_castellano))
```

```python
# Definir los datos
idiomas = ['Quechua', 'Aymara', 'Castellano']
cantidades = [len(i_quechua), len(i_aymara), len(i_castellano)]

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))
plt.bar(idiomas, cantidades, color=['blue', 'green', 'red'])
plt.title('Cantidad de Hablantes por Idioma')
plt.xlabel('Idioma')
plt.ylabel('Cantidad de Hablantes')
plt.show()
```

```python
i_noHabla = df_puno_juliaca[df_puno_juliaca['idioma_ninez']==7]
print("Hablantes de castellano: ",len(i_noHabla))
```

```python
varones = df_puno_juliaca[df_puno_juliaca['sexo']==1]
print("Cantidad de Varones: ",len(varones))

mujeres = df_puno_juliaca[df_puno_juliaca['sexo']==2]
print("Cantidad de Mujeres: ",len(mujeres))
```

```python
# Definir los datos
etiquetas = ['Varones', 'Mujeres']
cantidades = [len(varones), len(mujeres)]
total = sum(cantidades)

# Crear el gráfico de pastel
plt.figure(figsize=(8, 6))
plt.pie(cantidades, labels=etiquetas, autopct=lambda p: '{:.0f}'.format(p * total / 100), colors=['blue', 'pink'])
plt.title('Proporción de Varones y Mujeres en Puno y Juliaca')

# Agregar el total como texto
plt.text(-1.5, -1.5, f'Total: {total}', fontsize=12, bbox=dict(facecolor='white', alpha=0.5))

plt.show()
```

```python
# Filtrar los datos para Juliaca y Puno
juliaca = df[df['UBIGEO'] == 211101]
puno = df[df['UBIGEO'] == 210101]

# Contar la cantidad de varones y mujeres para Juliaca
juliaca_varones = juliaca[juliaca['sexo'] == 1].shape[0]
juliaca_mujeres = juliaca[juliaca['sexo'] == 2].shape[0]

# Contar la cantidad de varones y mujeres para Puno
puno_varones = puno[puno['sexo'] == 1].shape[0]
puno_mujeres = puno[puno['sexo'] == 2].shape[0]

# Mostrar la cantidad de registros de varones y mujeres en Juliaca y Puno
print("Juliaca:")
print("Varones:", juliaca_varones)
print("Mujeres:", juliaca_mujeres)
print()
print("Puno:")
print("Varones:", puno_varones)
print("Mujeres:", puno_mujeres)
```

```python
# Definir los datos
juliaca_cantidades = [94, 94]
puno_cantidades = [1234, 1380]
categorias = ['Varones', 'Mujeres']

# Crear el gráfico de barras
plt.figure(figsize=(8, 6))

plt.bar(categorias, juliaca_cantidades, color=['blue', 'blue'], width=0.4, label='Juliaca')
plt.bar([i + 0.4 for i in range(len(puno_cantidades))], puno_cantidades, color=['lightblue', 'lightblue'], width=0.4, label='Puno')

for i, valor in enumerate(juliaca_cantidades):
    plt.text(i, valor + 10, str(valor), ha='center', va='bottom')
for i, valor in enumerate(puno_cantidades):
    plt.text(i + 0.4, valor + 10, str(valor), ha='center', va='bottom')

plt.title('Cantidad de Varones y Mujeres en Juliaca y Puno')
plt.xlabel('Género')
plt.ylabel('Cantidad')
plt.xticks([i + 0.2 for i in range(len(categorias))], categorias)
plt.legend()
plt.show()
```
